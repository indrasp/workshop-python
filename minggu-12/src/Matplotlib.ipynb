```python
# Plotting dengan pengauran default 
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

```python
# Instansiasi default
import numpy as np
import matplotlib.pyplot as plt

# Buat gambar dengan ukuran 8x6 inci, 80 titik per inci
plt.figure (figsize = (8, 6), dpi = 80)

# Buat subplot baru dari kisi 1x1
plt.subplot (1, 1, 1)

X = np.linspace (-np.pi, np.pi, 256)
C, S = np.cos (X), np.sin (X)

# Plot cosinus dengan garis kontinu biru dengan lebar 1 (piksel)
plt.plot (X, C, color = "blue", linewidth = 1.0, linestyle = "-")

# Plot sinus dengan garis kontinu hijau dengan lebar 1 (piksel)
plt.plot (X, S, color = "green", linewidth = 1.0, linestyle = "-")

# Tetapkan batas x
plt.xlim (-4.0, 4.0)

# Atur x kutu
plt.xticks (np.linspace (-4, 4, 9))

# Tetapkan batas y
plt.ylim (-1.0, 1.0)

# Atur kutu y
plt.yticks (np.linspace (-1, 1, 5))

# Simpan gambar menggunakan 72 titik per inci
# plt.savefig ("exercise_2.png", dpi = 72)

# Tampilkan hasil di layar
plt.show ()
```

```python
# Regular Plots
n = 256
X = np.linspace(-np.pi, np.pi, n)
Y = np.sin(2 * X)

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.plot(X, Y - 1, color='blue', alpha=1.00)
```

```python
# Scatter Plots
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.scatter(X,Y)
```

```python
# Bar Plots
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

plt.ylim(-1.25, +1.25)
```

```python
# Contour Plots
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
```

```python
# Imshow
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 10
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)
plt.imshow(f(X, Y))
```

```python
# 3D Plots
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
